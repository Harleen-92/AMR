```python
df = pd.read_csv('Staphylococcus.pred.csv')
df = df.drop(columns=['study'], axis=1)
```

```python
df.shape
```

```python
antibiotics = df.columns[df.columns.str.contains('_i')]
print(antibiotics)
print(len(antibiotics))

```

```python
Results  = pd.DataFrame(columns=['Antibiotic','Methods','Accuracy','Sensitivity','specificity','AUROC',
                                 'AUPRC','N1','N2'])
```

```python
j=0
for i in range(10):
    df_pred = df.drop(columns=antibiotics,axis=1)
    response_antibiotic = antibiotics[i]
    print(response_antibiotic)
    df_pred[response_antibiotic] = df[response_antibiotic]
    print(Counter(df_pred[response_antibiotic]))
    df_pred[response_antibiotic]=df_pred[response_antibiotic].replace('Susceptible', 1)
    df_pred.loc[df_pred[response_antibiotic]!= 1, response_antibiotic] = 0
    label = df_pred.columns[0:9]
      ### Label Encoder ###                       
    labelEncoder = LabelEncoder()
    df_pred[label] = df_pred[label].apply(labelEncoder.fit_transform)
    X = df_pred.drop(columns=[response_antibiotic], axis=1)
    Y = df_pred[response_antibiotic]
                    ### Train Test Split
    x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.20)
    print(Counter(y_test))
#     print(x_train.shape)
#     print(x_test.shape)
#                             ##### Smote 
    ros = RandomOverSampler(random_state=123)
    rus = RandomUnderSampler(random_state=0)
    sm = SMOTE(random_state = 2) 
    sx_train, sy_train = sm.fit_sample(x_train, y_train)

                            ### Models 
    print( 'Start Random Forest Model')
    forest = RandomForestClassifier(random_state = 1)
    model = RandomForestClassifier()
    model.fit(sx_train, sy_train)
    S_RF_predictions = model.predict(x_test)
    S_RF_Accuracy = accuracy_score(y_test,S_RF_predictions)
    S_cm = confusion_matrix(y_test, S_RF_predictions)
    sensitivity = S_cm[0,0]/(S_cm[0,0]+S_cm[0,1])
    specificity = S_cm[1,1]/(S_cm[1,0]+S_cm[1,1])
    #print('Specificity : ', specificity)   
    auroc = roc_auc_score(y_test, S_RF_predictions)
    print(auroc)
    precision, recall, thresholds = precision_recall_curve(y_test, S_RF_predictions)
    auprc = auc(recall, precision)
    print(auprc)
    update = [response_antibiotic,'Random Forest',S_RF_Accuracy,sensitivity,specificity,
              auroc,auprc,Counter(y_test)[1],Counter(y_test)[0]]
    Results.loc[j] = update
    
    print('Feature Importance Start')
    r = permutation_importance(model, x_test,y_test,
                            n_repeats=5,
                            random_state=0)
    for i in r.importances_mean.argsort()[::-1]:
        if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
            print(f"{df_pred.columns[i]:<8}"
                  f"{r.importances_mean[i]:.3f}"
                  f" +/- {r.importances_std[i]:.3f}")
            
   
    sorted_idx = r.importances_mean.argsort()

    fig, ax = plt.subplots(figsize=(10,9))
    ax.boxplot(r.importances[sorted_idx].T,
           vert=False,labels=x_test.columns[sorted_idx],widths= 1)
    ax.set_title("Feature Importances for Susceptibilty Prediction of " + (response_antibiotic) )
    fig.tight_layout()
    plt.savefig(fname=response_antibiotic +'.png',dpi=300)
    plt.show()
    
    
    print('Random Forest Model Done')
    
#     print('SVM Start')
#     model = SVC()
#     model.fit(sx_train, sy_train)
#     S_SVM_predictions = model.predict(x_test)
#     auroc = roc_auc_score(y_test, S_SVM_predictions)
#     print(auroc)
#     precision, recall, thresholds = precision_recall_curve(y_test, S_SVM_predictions)
#     auprc = auc(recall, precision)
#     print(auprc)    
    
#     print('SVM Model Done')
    
    
    print('LR')
    model = LogisticRegression()
    model.fit(sx_train, sy_train)
    S_LR_predictions = model.predict(x_test)
    S_cm = confusion_matrix(y_test, S_LR_predictions)
    sensitivity = S_cm[0,0]/(S_cm[0,0]+S_cm[0,1])
    specificity = S_cm[1,1]/(S_cm[1,0]+S_cm[1,1])
    auroc = roc_auc_score(y_test, S_LR_predictions)
    print(auroc)
    S_LR_Accuracy = accuracy_score(y_test,S_LR_predictions)
    precision, recall, thresholds = precision_recall_curve(y_test, S_LR_predictions)
    auprc = auc(recall, precision)
    print(auprc)
    update = [response_antibiotic,'LR',S_RF_Accuracy,sensitivity,specificity,
              auroc,auprc,Counter(y_test)[1],Counter(y_test)[0]]
    Results.loc[j+1] = update
    print('Feature Importance Start')
#     r = permutation_importance(model, sx_train, sy_train,
#                             n_repeats=10,
#                             random_state=0)
#     for i in r.importances_mean.argsort()[::-1]:
#         if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
#             print(f"{df_pred.columns[i]:<8}"
#                   f"{r.importances_mean[i]:.3f}"
#                   f" +/- {r.importances_std[i]:.3f}")
#     sorted_idx = r.importances_mean.argsort()

#     fig, ax = plt.subplots()
#     ax.boxplot(r.importances[sorted_idx].T,
#            vert=False, labels=sx_train.columns[sorted_idx])
#     ax.set_title("Feature Importances for the LR Model on"+response_antibiotic)
#     fig.tight_layout()
#     plt.show()       
    print('LR Done')
    
    print('AdaBoost')
    
    model = AdaBoostClassifier()
    model.fit(sx_train, sy_train)
    S_AB_predictions = model.predict(x_test)
    S_AB_Accuracy = accuracy_score(y_test,S_AB_predictions)
    S_cm = confusion_matrix(y_test, S_AB_predictions)
    sensitivity = S_cm[0,0]/(S_cm[0,0]+S_cm[0,1])
    specificity = S_cm[1,1]/(S_cm[1,0]+S_cm[1,1])
    auroc = roc_auc_score(y_test, S_RF_predictions)
    print(auroc)
    precision, recall, thresholds = precision_recall_curve(y_test, S_RF_predictions)
    auprc = auc(recall, precision)
    print(auprc)
    update = [response_antibiotic,'AdaBoost',S_RF_Accuracy,sensitivity,specificity,
              auroc,auprc,Counter(y_test)[1],Counter(y_test)[0]]
    Results.loc[j+2] = update
    
    print('Feature Importance Start')
#     r = permutation_importance(model, sx_train, sy_train,
#                             n_repeats=10,
#                             random_state=0)
#     for i in r.importances_mean.argsort()[::-1]:
#         if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
#             print(f"{df_pred.columns[i]:<8}"
#                   f"{r.importances_mean[i]:.3f}"
#                   f" +/- {r.importances_std[i]:.3f}")
            
#     sorted_idx = r.importances_mean.argsort()

#     fig, ax = plt.subplots()
#     ax.boxplot(r.importances[sorted_idx].T,
#            vert=False, labels=sx_train.columns[sorted_idx])
#     ax.set_title("Feature Importances for the AdaBoost Model"+ (response_antibiotic))
#     fig.tight_layout()
#     plt.show()       
    
    print('AdaBoost Done')
    
    print('NaiveBayes Model')
    model = GaussianNB()
    model.fit(sx_train, sy_train)
    S_AB_predictions = model.predict(x_test)
    S_AB_Accuracy = accuracy_score(y_test,S_AB_predictions)
    S_cm = confusion_matrix(y_test, S_AB_predictions)
    sensitivity = S_cm[0,0]/(S_cm[0,0]+S_cm[0,1])
    specificity = S_cm[1,1]/(S_cm[1,0]+S_cm[1,1])
    auroc = roc_auc_score(y_test, S_AB_predictions)
    print(auroc)
    precision, recall, thresholds = precision_recall_curve(y_test, S_AB_predictions)
    auprc = auc(recall, precision)
    update = [response_antibiotic,'NaiveBayes',S_RF_Accuracy,sensitivity,specificity,auroc,
              auprc,Counter(y_test)[1],Counter(y_test)[0]]
    Results.loc[j+3] = update
    print(auprc)
    print('Feature Importance Start')
#     r = permutation_importance(model, sx_train, sy_train,
#                             n_repeats=10,
#                             random_state=0)
#     for i in r.importances_mean.argsort()[::-1]:
#         if r.importances_mean[i] - 2 * r.importances_std[i] > 0:
#             print(f"{df_pred.columns[i]:<8}"
#                f"{r.importances_mean[i]:.3f}"
#                f" +/- {r.importances_std[i]:.3f}")
#     sorted_idx = r.importances_mean.argsort()

#     fig, ax = plt.subplots()
#     ax.boxplot(r.importances[sorted_idx].T,
#            vert=False, labels=sx_train.columns[sorted_idx])
#     ax.set_title("Feature Importances for the NaiveBayes Model on"+response_antibiotic)
#     fig.tight_layout()
#     plt.show()       
    print('NaiveBayes Model Done')
    
    j=j+4
         
```
